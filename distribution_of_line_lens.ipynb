```python
import os
import pandas as pd
import shutil
os.chdir('/Users/BeckyMarcusMacbook/Thesis/manual_work/')
clean_text_dir = "/Users/BeckyMarcusMacbook/Thesis/TextCleaning/clean_text"
```

```python
rows = []
for root,dirs,files in os.walk(clean_text_dir):
    for file in files:
        if file[0]==".":
            continue
        group = os.path.split(root)[1]
        path = os.path.join(root,file)
        text = open(path).read()
        id = file.rsplit("-",1)[0]
        if len(text)==0:
            continue
        lines = [l for l in text.strip().splitlines() if len(l)!=0]
        wordses = [len(l.split()) for l in lines]
        lengths = [len(l) for l in lines]
        lpw = [l/w for w,l in zip(wordses,lengths)]
        rows.extend([(group,file,id,w,l,lp) for w,l,lp in zip(wordses,lengths,lpw)])
len(rows)
```

```python
df = pd.DataFrame(rows,columns = ["Group",'path',"id","words","length",'lpw'])
df = df.copy()[df['length']!=0]
```

```python
import re
dir = "clean_para_split"
for root,dirs,files in os.walk(clean_text_dir):
    for file in files:
        if file[0]==".":
            continue
        group = os.path.split(root)[1]
        path = os.path.join(root,file)
        text = open(path).read()
        id = file.rsplit("-",1)[0]
        lines = text.strip().splitlines()
        edited = "\n".join([l+"\n" if len(l)<20 else l for l in lines])
        dest = os.path.join(dir,group,file)
        os.makedirs(os.path.dirname(dest),exist_ok=True)
        with open(dest,'w') as f:
            f.write(edited)

```
